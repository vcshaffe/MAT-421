# 22.1 ODE Initial Value Problem Statement

A **diff. eq.** is a relationship between a function f(x) and its independent var., x. An **ordinary diff. eq.** is a diff. eq. wher the independent variable and therefore its derivatives, is in one dimension.

For the purpose of this course, we assume that an ODE can be written as a function F that incorporates one or all of the input arguments, which these input arguments are any nth order derivative of f(x), which we obtain an **nth order ODE.**

We can then generate a **general solution** to a diff. eq. that satisfies the equation. For an ODE of order n, a **particular solution** is a p(x) that satisfies the diff. eq. and n explicitly known values of the solution or its derivatives at certain points. 

# 22.2 Reduction of Order

Many of the methods used to solve IVTs are specifically meant for first order equations, so we often reduce the order of a diff. eq. to first order. We can often do this by putting equations and systems of equations into matrices and solving.

# 22.3 The Euler Method

Let dS(t)/dt = F(t,S(t)) be an explicitly defined first order ODE, such that F is a function that returns the derivative, and t be a numerical grid of the interval [t_0,t_f] with spacing h. Then, we can assume that t_0 = 0 and t_f = Nh for some positive int N.

The linear approximation of S(t) around t_j at t_(j+1) is S(t_(j+1)) = S(t_j) + (t_(j+1) - t_j) * dS(t_j)/dt, which can also be written as S(t_(j+1)) = S(t_j) + h*F(t_j, S(t_j)).

This formula is called the Explicit Euler Formula and allows us to compute an approximation for the state at S(t_(j+1)) given the state at S(t_j).

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

f = lambda t, s: np.exp(-t)
h = 0.05
t = np.arange(0, 1 + h,h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (8, 6))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approx, exact solutions')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

The explicit formula is called 'explicit' because it only requires information at t_j to compute the state at t_(j+1), while the Implicit Euler Formula can be derived by taking the linear approximation of S(t) around t_(j+1) and computing it at t_j:

S(t_(j+1)) = S(t_j) + h*F(t_(j+1), S(t_(j+1)))

There is also the Trapezoidal Formula which is the average of the explicit and implicit Euler formulas:

S(t_(j+1)) = S(t_j) + (h/2)*(F(t_j,S(t_j)) + F(t_(j+1),S(t_(j+1))))

Generally, only when an ODE is linear we can then use the Implicit or Trapezoidal formulas to approximate the solution to that equation.
